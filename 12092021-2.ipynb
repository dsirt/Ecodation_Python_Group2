```python
from tkinter import *
import random

liste = []

pencere = Tk()
pencere.title("UYGULAMA")
pencere.geometry("200x200+500+300")
#pencere.resizable(width = FALSE, height = FALSE)

etiket = Label(fg = "white", bg = "#61380B", font = "Helvetica 12 bold")
etiket.pack()

for i in range(6):
    while len(liste) != 6:
        a = random.randint(1, 100)
        if a not in liste:
            liste.append(a)
            
etiket["text"] = liste

mainloop()
```

```python
from tkinter import *
import random

liste = []

class Uygulama(object):
    def __init__(self):
        self.araclar()
        self.kodlar()
        
    def kodlar(self):
        for i in range(6):
            while len(liste) != 6:
                a = random.randint(1, 100)
                if a not in liste:
                    liste.append(a)  
        self.etiket["text"] = liste
        
    def araclar(self):
        self.etiket = Label(fg = "white", bg = "#61380B", font = "Helvetica 12 bold")
        self.etiket.pack()
        
pencere = Tk()
pencere.title("UYGULAMA")
pencere.geometry("200x200+500+300")
uyg = Uygulama()
mainloop()
```

```python
from tkinter import *

class Uygulama(object):
    def __init__(self):
        self.guiPenAr()
        
    def guiPenAr(self):
        self.etiket = Label(text = "Aşağıya e-posta adresinizi yazınız")
        self.etiket.pack()

        self.giris = Entry()
        self.giris.pack()

        self.cerceve = Frame()
        self.cerceve.pack(pady = 5)

        self.dugme = Button(text = "GÖNDER", command = pencere.destroy)
        self.dugme.pack()
        
pencere = Tk()
uyg = Uygulama()
mainloop()
```
