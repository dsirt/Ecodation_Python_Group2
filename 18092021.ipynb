```python
from tkinter import *

pencere = Tk()

v = IntVar()
v.set(0)

z = IntVar()
z.set(0)

d = StringVar()
d.set("Kubuntu")

e = StringVar()
e.set("Debian")


def onay():
    
    if v.get() == 1:
        giris.insert(0, "Merhaba %s kullanıcısı" % d.get())
        
    elif z.get() == 1:
        giris.delete(0, END)
        giris.insert(0, "Merhaba %s kullanıcısı" % e.get())
        
    else:
        giris.delete(0, END)
        
onay_pa = Checkbutton(text = d.get(), variable = v, command = onay)
onay_pa.place(relx = 0.0, rely = 0.1)

onay_de = Checkbutton(text = e.get(), variable = z, command = onay)
onay_de.place(relx = 0.0, rely = 0.2)

giris = Entry(width = 24)
giris.place(relx = 0.0, rely = 0.0)

mainloop()
```

```python
from tkinter import *

pencere = Tk()
pencere.geometry("400x200")

etiket = Label(text = "Farenin pencere üzerindeki konumu:")
etiket.pack(anchor = NW)

girisx_etiket = Label(text="X:")
girisx_etiket.pack(side = LEFT, anchor = N)

girisx = Entry(width = 5)
girisx.pack(side = LEFT, padx = 15, anchor = N)

girisy_etiket = Label(text="Y:")
girisy_etiket.pack(side = LEFT, anchor = N)

girisy = Entry(width = 5)
girisy.pack(side = LEFT, anchor = N)

def hareket(event = None):
    girisx.delete(0, END)
    girisy.delete(0, END)
    girisx.insert(0, event.x)
    girisy.insert(0, event.y)
    
pencere.bind("<Button-1>", hareket) #farenin sol tuşu, Button-2 -> farenin tekerleği, Button-3 -> farenin sağ tuşu

#Motion, B1-Motion, B2-motion, B3-Motion, ButtonRelease-1,-2,-3, Double-Button-1,-2,-3, Enter, Leave

mainloop()
```

```python
from tkinter import *

pencere = Tk()

pencere.geometry("200x200")

etiket = Label(text = "Basilan Tus:\n", wraplength = 150)
etiket.pack()

def goster(event = None):
    etiket["text"] += event.keysym
    
pencere.bind("<Key>", goster)

mainloop()
```

```python
from tkinter import *

pencere = Tk()
pencere.geometry("200x200")

giris = Entry()
giris.pack()

def silbakalım(event = None):
    if event.keysym == "BackSpace" or event.keysym =="Delete":
        return "break"
    
giris.bind("<Key>", silbakalım)

mainloop()
```
